```python
import numpy as np
from scipy import optimize
from scipy.optimize import root
import matplotlib.pyplot as plt
import math

x_min, x_max = -4, 5
y_min, y_max = -5, 4
step = 0.01

x, y = np.meshgrid(np.arange(x_min, x_max, step),
                   np.arange(y_min, y_max, step))

eq1 = 2 * x - np.cos(y + 1)
eq2 = y + np.sin(x) + 0.4

fig, ax = plt.subplots(figsize=(10, 10))

ax.contour(x, y, eq1, levels=[0], colors='red')

ax.contour(x, y, eq2, levels=[0], colors='blue')

ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('  ')
plt.grid(True)

plt.show()

x0 = 0.4
y0 = -1.5
delta = 0.1

def f1(y):
    return 2 * y + math.cos(y + 1)
def f2(x):
    return -0.4 - math.sin(x)

def iter(x, y, e):
    xn = x
    yn = y
    xn1 = f2(x)
    yn1 = f1(y)
    n = 1
    while ((abs(xn1 - xn) >= e) and (abs(yn1 - yn) >= e)):
        xn = xn1
        yn = yn1
        xn1 = f2(yn)
        yn1 = f1(xn)
        n += 1
    print('Simple iteration:')
    print('x =', xn, '\ny =', yn, '\nThe amount of iteration =', n)

iter(x0, y0, 0.0001)

def f3(x):
    eq1 = 2 * x[0] - math.cos(x[1] + 1)
    eq2 = x[1] + math.sin(x[0]) + 0.4
    return [eq1, eq2]

s = root(f3, [0.0, 0.0], method='hybr')
print('Check:', s.x)
```
