Метод Ейлера

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x + np.sin(y/np.sqrt(2))

a = 0.8
b = 1.8
h = 0.1
y0 = 1.3

n = int((b - a) / h)

x = np.linspace(a, b, n+1)

y = np.zeros(n+1)
y[0] = y0

for i in range(n):
    y[i+1] = y[i] + f(x[i], y[i]) * h

y_rounded=np.round_(y,4)
print("x=",x, "\ny=",y_rounded)

plt.plot(x,y, "o",x,y, "red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Метод Ейлера")
plt.legend(["точки", "x + np.sin(y/np.sqrt(2))"])
plt.grid()
plt.show()
```

Метод Ейлера-Коші

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x + np.cos(y/np.sqrt(3))

a = 1.2
b = 2.2
h = 0.1
y0 = 2.1
n = int((b - a) / h)

x = np.linspace(a, b, n+1)
y = np.zeros(n+1)
y[0] = y0

for i in range(n):
    y[i+1]=y[i]+(f(x[i],y[i])+f(x[i+1],y[i])+h*f(x[i],y[i]))*h/2

print("x =", x, "\ny =", y)
plt.plot(x,y, "o",x,y, "red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Метод Ейлера-Коші")
plt.legend(["точки", "x + np.cos(y/np.sqrt(3))"])
plt.grid()
plt.show()
```
